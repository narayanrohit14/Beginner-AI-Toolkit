# Set Up Data and Packages

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('citrus.csv')
```

```python
X = pd.get_dummies(df.drop(['name'], axis=1))
y = df['name'].apply(lambda x: 1 if x=='grapefruit' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
y_train.head()
```

# Import and Set Up Model

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

# Train and Predict with Model

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict([[2.96,86.76,172,85,2],[13.62,230.76,148,68,20]])
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
print(y_hat)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
print(y_hat)
```

```python
accuracy_score(y_test, y_hat)
```
